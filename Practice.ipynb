```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/Large_Dirty_Student_Dataset.csv")
```

```python
# df.head()
df.info()
# df.describe(include="all")
# df.shape
```

```python
#Duplicate Rows
df.duplicated().sum()
```

```python
#Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
#Check if duplicates removed or not
df.duplicated().sum()
```

```python
#Data Types
df.info()
```

```python
#Handle Data Types
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
```

```python
#Inconsistent Values
df["City"] = df["City"].str.strip().str.title()
```

```python
#check missing values
df.isnull().sum()
```

```python
#Handle missing values
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
df["Attendance"] = df["Attendance"].fillna(df["Attendance"].median())
```

```python
#Check invalid values (age smaller than 0)
df[df["Age"] < 0]
```

```python
#Removing age smaller than 0
df = df[df["Age"] >= 0]
```

```python
##Check invalid values (Marks greater than 100)
df[df["Marks"] > 100]
```

```python
#Check invalid values (attendence greater than 100)
df[df["Attendance"] > 100]
```

```python
#Remove or correct them
df = df[df["Marks"] <= 100]
df = df[df["Attendance"] <= 100]
```

```python
#Detect outliers

Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df["Marks"] < lower) | (df["Marks"] > upper)]
print(outliers)
```

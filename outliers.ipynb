```python
import pandas as pd
```

```python
dd = pd.read_csv("datasets/bad_daraz_dataset.csv")
dd
```

```python
dd["price"] = pd.to_numeric(dd["price"], errors = "coerce")
dd
```

```python
# Z-score method

mean = dd["price"].mean() #1853.84
std = dd["price"].std()  #2279.11

dd["z_score"] = (dd["price"] - mean)/std
dd

#z_score >  2 = Outliers
#z_score < 2  = Not-Outliers

outliers = dd[dd["z_score"].abs() > 2]
outliers
```

```python
#quartiles
median = dd["price"].quantile(0.50)
q1 = dd["price"].quantile(0.25)
q3 = dd["price"].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)  #-2550
upper_bound = q3 + (1.5 * iqr)   #8250

outliers = dd[(dd["price"] < lower_bound) | (dd["price"] > upper_bound)]
print(outliers)
```

```python
len(outliers)
```

```python
#fixing outliers using Mean/Median
dd.loc[(dd["price"] < lower_bound) | (dd["price"] > upper_bound), "price"] = median
dd
```

```python
#fixing outliers using winsorization(capping) ---> clip 
dd["price"].clip(lower_bound,upper_bound)
```

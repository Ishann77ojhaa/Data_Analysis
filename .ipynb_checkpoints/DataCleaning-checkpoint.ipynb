```python
bdd = pd.read_csv("datasets/bad_daraz_dataset.csv")
bdd
```

```python
dummy = pd.Series(["100","200","Ishan"])
pd.to_numeric(dummy, errors = "coerce")
```

```python
# bdd.isnull().sum()  #Detects NAN
# bdd.fillna(0)     #Puts NAN as 0

#cleaning price
bdd["price"] = pd.to_numeric(bdd["price"], errors = "coerce")
bdd
```

```python
bdd["price"] = bdd["price"].fillna(bdd["price"].mean()) #fill NAN with mean of all
bdd
```

```python
#cleaning age
bdd["age"] = pd.to_numeric(bdd["age"], errors = "coerce")
bdd
```

```python
bdd["age"] = bdd["age"].fillna(bdd["age"].median())  #fill NAN with median of all
bdd
```

```python
bdd.dropna() #Removes all rows which has NAN
```

```python
bdd[bdd.duplicated()]  #finding duplicate row
```

```python
bdd["discount_percent"] = bdd["discount_percent"].str.replace("%", "")  #Removing % in one of the data in discount_percent
bdd
```

```python
bdd
```

```python
bdd.drop_duplicates(inplace = True) #Remove complete Duplicates 
```

```python
#for removing partial duplications
bdd.drop_duplicates(subset= "product_id", inplace = True) 
#if product id repeats remove that , always which comes first is origional and second one will be removed. (Default)
```

```python
bdd
```

```python
bdd.drop_duplicates(subset = "product_id", keep = "last", inplace = True) #Removes the first one and keeps the 2nd one
```

```python
bdd
```

```python
bdd["price"] = pd.to_numeric(bdd["price"], errors = "coerce")  #String into NAN
bdd
```

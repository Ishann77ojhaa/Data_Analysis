```python
lms[student_name] = lms[student_name].str.title()
lms[course] = lms[course].str.title()
lms[city] = lms[city].str.title()
lms[mentor] = lms[mentor].str.title()
```

```python
lms["age"] = lms.loc[lms['age'] < 0, "age"] = abs(lms["age"])
```

```python
columns_forIQR = [
    "course_fee",
    "attendance_percent",
    "assignment_score",
    "final_score"
]

for columns in columns_forIQR:
    Q1 = lms[columns].quantile(0.25)
    Q3 = lms[columns].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = lms[(lms[columns] < lower) | (lms[columns] > upper)]

print(outliers[[columns]])
```

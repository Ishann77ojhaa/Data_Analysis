```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fruits = ['Apple', 'Banana', 'Orange']
sales = [100, 150, 80]
revenue = [20,30,40]

#Line chart
plt.Figure(figsize=(6,8)) #pixels 
plt.plot(sales, revenue,c = 'red', marker='o')
plt.xlabel("Sales")
plt.ylabel("Revenue")
plt.title("Fruit Sales")
plt.show()
```

```python
dd = pd.read_csv("datasets/daraz_dataset.csv")
```

```python
category = dd.groupby("Category")["Product"].count()
category.index
category.values
```

```python
#Bar chart

plt.figure(figsize=(10,6))
plt.bar(category.index, category.values, color = "green", edgecolor = "black")
plt.xlabel("Categories", fontsize = 20)
plt.ylabel("Total Products", fontsize = 20)
plt.title("Products according to category",fontsize = 20)
plt.grid(axis = "y", linestyle = "--", alpha = 0.5)
plt.show()
```

```python
category
```

```python
#Pie Chart

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0' , 'grey', 'yellow']  #for custom colors
explode = (0.1, 0, 0, 0, 0,0,0) #for little highlight
plt.figure(figsize=(12, 6))  #for increasing size
plt.pie(category.values, labels = category.index, colors = colors, shadow = True, explode=explode)
plt.title("Category Breakdown")
plt.show()
```

```python
gender = dd.groupby("Gender")["Product"].count()
gender
```

```python
plt.pie(gender.values, labels = gender.index, shadow = True)
plt.title("Clssification acc to gender")
```

```python
#Histogram

plt.hist(dd["UnitPrice"], color = "skyblue", edgecolor = "black", alpha = 0.8, bins = 8)  #arguments are all same as bar graph

# Add labels and title
plt.xlabel('Unit-Price')
plt.ylabel('Frequency')
plt.title('Basic Histogram Example')

# Display the plot
plt.show()
```
